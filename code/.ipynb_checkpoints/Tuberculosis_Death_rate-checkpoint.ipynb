1. Manually calculate mean:

    Write a function that takes in values and returns the mean of the values. Create a list of numbers that you test on your function to check to make sure your function works!
    
    *Note*: Do not use any mean methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
def avg(numbers):
    tot = 0
    count = 0
    for i in numbers:
        tot += i
        count += 1
    avg = tot/count
    return avg

avg([5, 10, 15])
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
from math import *
```

```python
def std(numbers):
    avg_num = avg(numbers)
    sum_sqrd_diff = sum((i - avg_num) ** 2 for i in numbers)
    variance = sum_sqrd_diff / len(numbers)
    std_dev = sqrt(variance)
    
    return std_dev
```

```python
def clean_data(data):
    cleaned_data = []
    for i in data:
        if any(c.isdigit() for c in i):
            cleaned_item = ''.join(c if c.isdigit() else '' for c in i)
            cleaned_data.append(int(cleaned_item))
        else:
            cleaned_data.append(int(i))
    return cleaned_data

clean_data(['45K'])
```

```python
#inspiration on how to merge was used from https://pandas.pydata.org/docs/user_guide/merging.html
pop_df = pd.DataFrame(pop)
le_df = pd.DataFrame(le)
tb_df = pd.DataFrame(tb)

final_df = pd.merge(pop_df, le_df, on='country', how='inner', suffixes=('_pop', '_le'))
final_df = pd.merge(final_df, tb_df, on='country', how='inner')
final_df.head(5)
```

```python
final_df.info()
```

```python
def clean_columns(value):
    if isinstance(value, str):
        if 'M' in value or 'm' in value:
            return float(value.replace('M', '').replace('m', '')) * 1_000_000  # Convert millions to numeric
        elif 'K' in value or 'k' in value:
            return float(value.replace('K', '').replace('k', '')) * 1_000  # Convert thousands to numeric
        elif 'B' in value or 'b' in value:
            return float(value.replace('B', '').replace('b', '')) * 1_000_000_000  # Convert billions to numeric
        else:
            return float(value.replace(',', '').replace('$', ''))
    else:
        return value
```

```python
pop_columns = final_df.filter(like='_pop').columns
final_df[pop_columns] = final_df[pop_columns].applymap(clean_columns)
```
